## Before Scalling as We have some non numeric columns, So we have to encode them
# Encoding

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

Le=LabelEncoder()
OneHot=OneHotEncoder()

df["Gender"]=Le.fit_transform(df["Gender"])
df["Extracurricular_Activities"]=Le.fit_transform(df["Extracurricular_Activities"])
df["Internet_Access_at_Home"]=Le.fit_transform(df["Internet_Access_at_Home"])
df["Grade"]=Le.fit_transform(df["Grade"])
df["Family_Income_Level"]=Le.fit_transform(df["Family_Income_Level"])



```

```python
df=pd.get_dummies(df,columns=["Department","Parent_Education_Level"], dtype=int)

```

```python
df.head()
```

# 3. Feature Scaling

## Apply both:

### StandardScaler

### MinMaxScaler

```python
df.head()

X=df.drop("Grade",axis=1)
y=df["Grade"]
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler

Sc=StandardScaler()
MinMax=MinMaxScaler()

X_Standard=X.copy()
X_MinMax=X.copy()

cols=list( ["Age",
    "Attendance (%)",
    "Midterm_Score",
    "Final_Score",
    "Assignments_Avg",
    "Quizzes_Avg",
    "Participation_Score",
    "Projects_Score",
    "Study_Hours_per_Week",
    "Stress_Level (1-10)",
    "Sleep_Hours_per_Night"])


X_Standard[cols]=Sc.fit_transform(X_Standard[cols])
X_MinMax[cols]=Sc.fit_transform(X_MinMax[cols])
```

# 4. Data Splitting

## Split the dataset into:

### Training Set

### Testing Set

# Data spliting on the basis of StandardScaler

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X_Standard,y,test_size=0.2,random_state=42)

```

# 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score


Logistic_model=LogisticRegression()
Logistic_model.fit(x_train,y_train)
y_predicted=Logistic_model.predict(x_test)

Logistic_Accuracy=accuracy_score(y_test,y_predicted)
Logistic_F1_Score=f1_score(y_test,y_predicted,average="weighted")

print("Accuracy score : ",(Logistic_Accuracy*100))
print("F1 score : ",(Logistic_F1_Score*100))

```

# 2. KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier

KNN_model=KNeighborsClassifier(n_neighbors=3)
KNN_model.fit(x_train,y_train)
y_predicted=KNN_model.predict(x_test)

KNN_Accuracy=accuracy_score(y_test,y_predicted)
KNN_F1_Score=f1_score(y_test,y_predicted,average="weighted")

print("Accuracy score : ",(KNN_Accuracy *100))
print("F1 score : ",(KNN_F1_Score*100))
```

# 3. Decision Tree


```python
from sklearn.tree import DecisionTreeClassifier

Decision_model=LogisticRegression()
Decision_model.fit(x_train,y_train)
y_predicted=Decision_model.predict(x_test)

DecisionTree_Accuracy=accuracy_score(y_test,y_predicted)
DecisionTree_F1_Score=f1_score(y_test,y_predicted,average="weighted")

print("Accuracy score : ",(DecisionTree_Accuracy*100))
print("F1  score : ",(DecisionTree_F1_Score*100))
```

# 4. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
RandForest_model=RandomForestClassifier(random_state=42)

RandForest_model.fit(x_train,y_train)
y_predicted=RandForest_model.predict(x_test)

RandomForest_Accuracy=accuracy_score(y_test,y_predicted)
RandomForest_F1_Score=f1_score(y_test,y_predicted,average="weighted")

print("Accuracy Score : ",RandomForest_Accuracy*100)
print("F1 Score : ",RandomForest_F1_Score*100)
```

```python
result=pd.DataFrame({
    'Model':["Logistic Regression","KNN","Decision Tree","Random Forest"],
    'Accuracy Score':[Logistic_Accuracy,KNN_Accuracy,DecisionTree_Accuracy,RandomForest_Accuracy],
    'F1 Score':[Logistic_F1_Score,KNN_F1_Score,DecisionTree_F1_Score,RandomForest_F1_Score]
})


result["Accuracy Score"]=result["Accuracy Score"]*100
result["F1 Score"]=result["F1 Score"]*100

```

```python
result.set_index("Model").plot(kind="bar")
plt.xlabel("Model")
plt.ylabel("Accuracy and F1 Score ")
plt.title("Performance Comparison Table (Standard Scaler)")
plt.show()
```
